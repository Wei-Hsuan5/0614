```python
#下載中文字型
import wget
wget.download("https://github.com/roberthsu2003/machine_learning/raw/refs/heads/main/source_data/ChineseFont.ttf")
```

```python
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")
```

```python
import math
import numpy as np
from numpy import ndarray
import matplotlib.pyplot as plt

t:ndarray=np.arange(0, 2.5, 0.1)
#t.ndim
#t.size
#t.shape
#t.dtype
#print(t)
y1:ndarray=np.sin(t*math.pi)
y2:ndarray=np.sin(t*math.pi+math.pi/2)
y3:ndarray=np.sin(t*math.pi-math.pi/2)
figure, axes=plt.subplots()
axes.plot(t, y1, marker='s')
axes.plot(t, y2, marker='*', linestyle='-.')
axes.plot(t, y3, marker='+', linestyle='--')
axes.legend(loc='upper left', bbox_to_anchor=(-0.1,1))
axes.set_title("sin()的正弦值")
axes.set_xlabel("圓周率")
axes.set_ylabel("正弦波值")
plt.show()
```

```python
#下載中文字型
import wget
wget.download("https://github.com/roberthsu2003/machine_learning/raw/refs/heads/main/source_data/ChineseFont.ttf")
```

```python
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")
```

```python
import math
import numpy as np
from numpy import ndarray
import matplotlib.pyplot as plt

t:ndarray=np.arange(0, 2.5, 0.1)
#t.ndim
#t.size
#t.shape
#t.dtype
#print(t)
y1:ndarray=np.sin(t*math.pi)
y2:ndarray=np.sin(t*math.pi+math.pi/2)
y3:ndarray=np.sin(t*math.pi-math.pi/2)

plt.plot(t, y1, marker='s', linestyle='', label='90')
plt.plot(t, y2, marker='*', linestyle='-.', label='0')
plt.plot(t, y3, marker='+', linestyle='--', label='-90')
plt.legend(loc='upper right', bbox_to_anchor=(1.18,1))
plt.title("sin()的正弦值")
plt.xlabel("圓周率")
plt.ylabel("正弦波值")
plt.show()
```
