**製作圓餅圖**

```python
import matplotlib.pyplot as plt

labels = ['Cocktails', 'Beer', 'Soju', 'Spirits']
values = [15, 25, 40, 10]
colors = ['pink', 'green', 'teal', 'gray']
explode = [0, 0, 0.1, 0]

_, _, autotexts=plt.pie(values, labels=labels, colors=colors, explode=explode, shadow= True, startangle=-30, autopct='%1.1f%%')
plt.setp(autotexts, size=10, color='white')
plt.show()
```
